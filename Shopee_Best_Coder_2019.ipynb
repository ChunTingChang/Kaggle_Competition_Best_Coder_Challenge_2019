```python
import pandas as pd
import numpy as np
from datetime import datetime

import re
from glob import glob

pd.options.mode.chained_assignment = None
```

```python
# get the data for submission and keywords group

paths = glob('./res/*.csv')
df_subs = pd.read_csv(paths[0]) # keywords group
df_ques = pd.read_csv(paths[1]) # for submission
```

```python
display(df_ques.info(), 
        df_subs.info())
```

```python
display(df_ques.head(), df_subs.head(10))
```

```python
# replace symbols

def only_letters(words):
    # keep only numbers and letters and spaces
    clean = re.sub('[^0-9 a-z A-Z]', '', words).strip().lower()
    clean_list = clean.split(" ")
    return clean

df_ques['clean_words'] = df_ques['name'].apply(only_letters)
```

```python
# take a look at the clean result

df_ques['groups_found'] = ''
display(df_ques.clean_words.values[40:50], df_ques.head())
```

```python
# replace `-` and make keywords a list in cell

def keyword_group(words):
    # make keywords a list in cell
    group_items = words.replace("-", "").split(', ')
    return group_items

def keyword_cnt(words):
    # counts of keywords in the list
    group_items = words.replace("-", "").split(', ')
    num_of_items = len(group_items)
    return num_of_items

df_subs['kw_group'] = df_subs.Keywords.apply(keyword_group)
df_subs['kw_cnt'] = df_subs.Keywords.apply(keyword_cnt)
```

```python
display(df_subs.head(),
df_ques.head())
```

```python
# function to check if kw in the provided title text

def match_kw(test_str):
    output = []
    for kw, num, grp in zip(df_subs.kw_group, df_subs.kw_cnt, df_subs.Group):
        
        # if only 1 keyword for 1 group
        if num == 1:
            if kw[0] in test_str:
                output.append(grp)
                
        # if more than 1 keywords for 1 group
        else:
            for iteration in range(num):
                if kw[iteration] in test_str:
                    output.append(grp)
    return output

df_ques['groups_found'] = df_ques['clean_words'].apply(match_kw)
```
